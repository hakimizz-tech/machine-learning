```python
##DDA Agorithim
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')


def dda(x0, y0, x1, y1):
  # find the delta of x and y
  dx = x1 - x0
  dy = y1 - y0

  # find the steps
  steps = max(dx, dy)

  # find the y increment and x increment
  y_incr = abs(dy/steps)
  x_incr = abs(dx/steps)

  # take the y0  and x0 as the starting point
  y = y0
  x = x0

  y_cordinates = []
  x_cordinates = []

  for i in range(steps):
    x += x_incr
    y += y_incr

    # increment the values
    y_cordinates.append(y)
    x_cordinates.append(x)

  plt.plot(x_cordinates, y_cordinates, marker='o', markersize=1, markerfacecolor="green")
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.title('Digital Differential Analyzer')
  plt.legend('Pixels')
  plt.show()
```

```python
# coordinates of 1st point
x0, y0 = 20, 20

# coordinates of 2nd point
x1, y1 = 60, 50

# Function call
dda(x0, y0, x1, y1)
```
